```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = {
    "student":["Alex","Joy","Sam","Bri","Ali","Janet"],
    "course":["Databases","Python","AI","Machine learning","Datascience","Forensics"],
    "score":[78,85,90,88,75,82]
}
df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.describe()
```

```python
#get avg score per student
student_avg = df.groupby("student")["score"].mean()
student_avg
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
student_avg.plot(kind="bar")
plt.title("Average score per student")
plt.xlabel("Student")
plt.ylabel("Average score")
plt.show()
```

```python
#analysing how each course is performing
course_avg = df.groupby("course")["score"].mean()
course_avg
```

```python
course_avg.plot(kind="bar")
plt.title("Average course performance")
plt.xlabel("course")
plt.ylabel("average score")
plt.show()
```
